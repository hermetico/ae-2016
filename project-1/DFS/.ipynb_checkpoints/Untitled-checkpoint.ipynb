```python

# -*- coding: utf-8 -*-
import math
LENGTH = 5
HEIGHT = int(math.floor( math.log(LENGTH, 2) ) ) + 1

data  = [ x for x in range(1, LENGTH + 1)]
tree  = [0] if HEIGHT == 1 else [0] * (2 ** HEIGHT) # == 2‚Åµ

def fillDFSRecursive(tree, data, init, end, v, index = 0, branch='root', parent=''):
    """ Recursive part """
    
    # I guess when we are at the end of the tree, we just quit
    if v == 0: return
    
       
    size = end - init
    pointer =  init + int(size * 0.5)
    

    ## DEBUG PRINT
    print """    -----
    init   : %s end: %s v: %s index: %s size: %s pointer: %s 
    type index   :%s
    type pointer : %s
    -------
    """ %(init, end, v, index, size, pointer, type(index), type(pointer))

    
    
    tree[index] = (parent, branch, data[pointer])
    ## next steps

    if size == 0: return
    # left branch
    fillDFSRecursive(tree, data, init, pointer, v - 1, index + 1, 'left', data[pointer])
       
    fillDFSRecursive(tree, data, pointer + 1, end, v - 1, int(index + math.pow(2, v - 1)), 'right', data[pointer])
    
    
    
def fillDFS(tree, data):
    """ Main function for DFS """
    # ceil rounds to the nex tint
    height =  HEIGHT
    fillDFSRecursive(tree, data, 0, len(data) - 1, height)
```

```python
fillDFS(tree, data)
print data
print tree
```
